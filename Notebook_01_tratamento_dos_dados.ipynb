###2.2. Ajuste e padronização das features

```python
# Adicionando o ano de referência

df_2022['ano'] = 2022
df_2023['ano'] = 2023
df_2024['ano'] = 2024

# ---------------------------------------------------------
# PADRONIZAÇÃO DO DATASET DE 2022
# ---------------------------------------------------------
df_2022_tratado = df_2022.rename(columns={
    'RA': 'ra',                                           # Registro do Aluno
    'Nome': 'nome',                                       # Nome do Aluno (Dados estão Anonimizados)
    'Idade 22': 'idade',                                  # Idade do Aluno
    'Gênero': 'genero',                                   # Gênero do aluno
    'Fase': 'fase',                                       # Fase é o Nível de Aprendizado do Aluno
    'Turma': 'turma',                                     # Número da Turma de cada fase (1A, 1B)
    'Ano ingresso': 'ano_ingresso',                       # Ano que o Aluno Ingressou na Passos Mágicos
    'Instituição de ensino': 'instituicao_ensino',        # Instituição de Ensino do Aluno
    'Rec Psicologia': 'rec_psicologia',
    'Indicado': 'indicado',

    # Indicadores e Métricas
    'INDE 22': 'inde',                                    # Índice do Desenvolvimento Educacional (Métrica Geral)
    'Pedra 22': 'pedra',                                  # Classificação do Aluno (Quartzo, Ágata, Ametista, Topázio)
    'IAA': 'iaa',                                         # Indicador de Auto Avaliação
    'IEG': 'ieg',                                         # Indicador de Engajamento
    'IPS': 'ips',                                         # Indicador Psicossocial
    'IDA': 'ida',                                         # Indicador de Aprendizagem
    'IPV': 'ipv',                                         # Indicador de Ponto de Virada
    'IAN': 'ian',                                         # Indicador de Adequação ao Nível

    # Notas
    'Matem': 'nota_mat',                                  # Média das Notas de Matemática
    'Portug': 'nota_por',                                 # Média das Notas de Português
    'Inglês': 'nota_ing',                                 # Média das Notas de Inglês

    # Status
    'Atingiu PV': 'atingiu_pv',                           # Booleano que sinaliza se o Aluno atingiu o Ponto de Virada
    'Fase ideal': 'fase_ideal',                           # Nível (Fase) ideal do Aluno
    'Defas': 'defasagem'                                  # Nível de defasagem do aluno
})


# ---------------------------------------------------------
# PADRONIZAÇÃO DO DATASET DE 2023
# ---------------------------------------------------------
df_2023_tratado = df_2023.rename(columns={
    'RA': 'ra',
    'Nome Anonimizado': 'nome',
    'Idade': 'idade',
    'Gênero': 'genero',
    'Fase': 'fase',
    'Turma': 'turma',
    'Ano ingresso': 'ano_ingresso',
    'Instituição de ensino': 'instituicao_ensino',
    'Rec Psicologia': 'rec_psicologia',
    'Indicado': 'indicado',

    'INDE 2023': 'inde',
    'Pedra 2023': 'pedra',
    'IAA': 'iaa',
    'IEG': 'ieg',
    'IPS': 'ips',
    'IDA': 'ida',
    'IPV': 'ipv',
    'IAN': 'ian',
    'IPP': 'ipp',                                         # Indicador Psicopedagógico (Inserido em 2023)

    'Mat': 'nota_mat',
    'Por': 'nota_por',
    'Ing': 'nota_ing',

    'Atingiu PV': 'atingiu_pv',
    'Fase Ideal': 'fase_ideal',
    'Defasagem': 'defasagem'
})

# ---------------------------------------------------------
# PADRONIZAÇÃO DO DATASET DE 2024
# ---------------------------------------------------------
df_2024_tratado = df_2024.rename(columns={
    'RA': 'ra',
    'Nome Anonimizado': 'nome',
    'Idade': 'idade',
    'Gênero': 'genero',
    'Fase': 'fase',
    'Turma': 'turma',
    'Ano ingresso': 'ano_ingresso',
    'Instituição de ensino': 'instituicao_ensino',
    'Rec Psicologia': 'rec_psicologia',
    'Indicado': 'indicado',

    'INDE 2024': 'inde',
    'Pedra 2024': 'pedra',
    'IAA': 'iaa',
    'IEG': 'ieg',
    'IPS': 'ips',
    'IDA': 'ida',
    'IPV': 'ipv',
    'IAN': 'ian',
    'IPP': 'ipp',

    'Mat': 'nota_mat',
    'Por': 'nota_por',
    'Ing': 'nota_ing',

    'Fase Ideal': 'fase_ideal',
    'Defasagem': 'defasagem'
})
```

### 2.3. Filtros e Concatenação

```python

# Listando as colunas finais

colunas_finais = [
    'ano', 'ra', 'nome', 'idade', 'genero', 'fase', 'turma', 'ano_ingresso',
    'instituicao_ensino', 'rec_psicologia', 'indicado',
    'inde', 'pedra', 'fase_ideal', 'defasagem', 'atingiu_pv',
    'iaa', 'ieg', 'ips', 'ida', 'ipv', 'ian', 'ipp',
    'nota_mat', 'nota_por', 'nota_ing'
]

# Filtrando os data-frames

df_22_filtrado = df_2022_tratado[[col for col in colunas_finais if col in df_2022_tratado.columns]]
df_23_filtrado = df_2023_tratado[[col for col in colunas_finais if col in df_2023_tratado.columns]]
df_24_filtrado = df_2024_tratado[[col for col in colunas_finais if col in df_2024_tratado.columns]]

# Juntando as colunas para o dataframe final

df_final = pd.concat([df_22_filtrado, df_23_filtrado, df_24_filtrado], ignore_index=True)
```

### 2.4 Realizando ajustes no df_final

```python
# Ajustando colunas númericas decimais (Dtype "object")
colunas_numericas = [
    'inde', 'iaa', 'ieg', 'ips', 'ida', 'ipv', 'ian', 'ipp',
    'nota_mat', 'nota_por', 'nota_ing'
]

# Substituindo vírgula (,) por ponto (.) e células vazias
for col in colunas_numericas:
    if col in df_final.columns:
      if df_final[col].dtype == 'object':
        df_final[col] = df_final[col].str.replace(',', '.', regex=False)
      df_final[col] = pd.to_numeric(df_final[col], errors='coerce')
```

```python
# Convertendo Idade, Fase e Ano de Ingresso para inteiros
colunas_inteiras = ['idade', 'ano_ingresso', 'fase']
for col in colunas_inteiras:
    if col in df_final.columns:
        if df_final[col].dtype == 'object':
            df_final[col] = df_final[col].astype(str).str.extract(r'(\d+)')[0]
        df_final[col] = pd.to_numeric(df_final[col], errors='coerce').astype('Int64')
```

### 2.5 Verificação de conteúdo das features e realizando ajustes

**Numéricas:** idade (anos), ano_ingresso, notas acadêmicas (nota_mat, nota_por – escala de 0 a 10) e os indicadores de desempenho da instituição (inde, iaa, ieg, ips, ida, ipv, ian, ipp – escala de 0 a 10). *Observação:* o dataset original contém inconsistências de formatação; no tratamento, os tipos foram convertidos para float e os dados faltantes foram imputados utilizando a mediana.

**Ordinais (escalas):** fase (nível de aprendizado do aluno na ONG, ex: Alfa, Fase 1 até Fase 8), defasagem (grau de atraso/avanço escolar) e pedra (classificação evolutiva de desempenho: Quartzo, Ágata, Ametista, Topázio). *Observação:* os textos da coluna pedra foram padronizados para corrigir divergências de digitação e acentuação.

**Categóricas:** genero (Masculino, Feminino), turma e fase_ideal. (Identificadores únicos excluídos da modelagem: ra e nome).

```python
df_definitivo = df_final.copy()

# Removendo colunas desnecessárias e linhas vazias
colunas_para_remover = ['atingiu_pv', 'nota_ing','nota_mat','nota_por','rec_psicologia', 'indicado',]
df_definitivo = df_definitivo.drop(columns=[col for col in colunas_para_remover if col in df_definitivo.columns])
df_definitivo = df_definitivo.dropna(subset=['ra', 'pedra', 'fase'])

# Preenchendo valores nulos (NaN) das notas e indicadores com a MEDIANA
colunas_para_preencher = [
    'inde', 'iaa', 'ieg', 'ips', 'ida', 'ipv', 'ian', 'ipp'
]
for col in colunas_para_preencher:
    if col in df_definitivo.columns:
        df_definitivo[col] = df_definitivo[col].fillna(df_definitivo[col].median())

# Limpando espaços e aplicando 1ª letra Maiúscula
colunas_texto = ['genero', 'pedra', 'fase_ideal', 'instituicao_ensino']
for col in colunas_texto:
    if col in df_definitivo.columns:
        df_definitivo[col] = df_definitivo[col].astype(str).str.strip().str.title()
        df_definitivo[col] = df_definitivo[col].replace('Nan', np.nan)

# Substituindo nomes divergentes para um padrão único
df_definitivo['genero'] = df_definitivo['genero'].replace({'Menino': 'Masculino', 'Menina': 'Feminino'})
df_definitivo['pedra'] = df_definitivo['pedra'].replace({'Agata': 'Ágata', 'Topazio': 'Topázio'})

# Transformando a 'fase_ideal' em números (0 a 8)
fases_numericas = df_definitivo['fase_ideal'].astype(str).str.extract(r'(?i)fase\s*(\d+)')[0]

df_definitivo['fase_ideal'] = fases_numericas.fillna(0).astype(int)
```

```python
# Binarizando a coluna Instituição de Ensino

def classificar_escola(nome_escola):
    nome = str(nome_escola).lower()

    privadas = ['privada', 'particular', 'decisão', 'decisao', 'jp ii', 'joão paulo', 'joao paulo']
    publicas = ['pública', 'publica', 'estadual', 'municipal'] # Adicionei estadual/municipal que significam pública

    if any(palavra in nome for palavra in privadas):
        return 'Privada'
    elif any(palavra in nome for palavra in publicas):
        return 'Pública'
    else:
        return 'Não Classificada'

# Aplicando a função na coluna
df_definitivo['instituicao_ensino'] = df_definitivo['instituicao_ensino'].apply(classificar_escola)

# Mantendo apenas as pedras válidas
pedras_validas = ['Quartzo', 'Ametista', 'Ágata', 'Topázio']
df_definitivo = df_definitivo[df_definitivo['pedra'].isin(pedras_validas)].copy()
```

## 4. Criando a variável Alvo (TARGET) do Modelo

```python
# Risco = 1 (Quartzo ou Ágata) | Risco = 0 (Ametista ou Topázio)
df_definitivo['risco_defasagem'] = np.where(df_definitivo['pedra'].isin(['Quartzo', 'Ágata']), 1, 0)

# CONFERÊNCIA

print("--- Valores Únicos após a Padronização ---")
print(f"Gênero: {df_definitivo['genero'].unique()}")
print(f"Pedra: {df_definitivo['pedra'].unique()}")

print("\n--- Distribuição da Variável Target (Risco) ---")
print(df_definitivo['risco_defasagem'].value_counts())

print("\n--- Valores Nulos Restantes ---")
print(df_definitivo.isnull().sum())
```
